```python
#SECTION 1) LAUNCH PROJECT
#welcome banner

project = ("Open Track", 2.0 , 2026 , "Python", "Siti Aisyah")

print("=" * 40)
print(f"{project[0]} - Open Source Project")
print(f"Project Lead : {project[4]}")
print("=" * 40)
print(f"Name : {project[0]} | Version : {project[1]} | Started : {project[2]} ")
print(f"First 3 fields : {project[0:3]}")
print(f"Language count : {project.count(project[3])}   Language index : {project.index("Python")}")
#project[0] = 'test' ......
print("-" * 40)

contributors = []

for i in range(4):
  print(f"Contributor {i+1}")
  name = input("Name     : ")
  role = input("Role     : ")
  language = input("Language : ")
  commits = input("Commits  : ")
  country = input("Country  : ")
  print("-" * 40)

  contributor = {
      "name" : name,
      "role" : role,
      "language" : language,
      "commits" : commits,
      "country" : country
  }

  contributors.append(contributor)


#extract names
names = [c["name"] for c in contributors]
names.sort()

print("Sorted names : ", names)

#last name(negative index)  / slice first 2 names
print(f"Last name : {names[-1]}   First two : {names[:2]}")

#add status using update()
for c in contributors:
  c.update({"status" : "Active"})


#print the 1st contributor status using get
print(f"Contributor 1 status : {contributors[0].get('status')}")

#backup contributor using copy()
backup = contributors[0].copy()
print(f"Backup : {backup}")
print("-" * 40)
print(" ")



#SECTION 2) TRACK AND ANALYSE ISSUES

issues = []

for i in range(5):
  print(f"Issue {i+1}")
  id = input("ID       : ")
  title = input("Title    : ")
  types = input("Type (Bug/Feature): ")
  priority = input("Priority (Critical/High/Medium/Low): ")
  reporter = input("Reporter : ")
  status = input("Status (Open/In Progress/Resolved) : ")
  print('-' * 40)
  issue = {
      "id" : id,
      "title" : title,
      "types" : types,
      "priority" : priority,
      "reporter" : reporter,
      "status" : status
  }

  issues.append(issue)

#manually count Open
open_count = 0

for issue in issues:
  if issue["status"] == "Open":
    open_count = open_count + 1

print(f"\nOpen issues : {open_count}")


#change first priority
issues[0]["priority"] = "Critical"
print("First issue > priority updated to Critical.")

#slice last 2
print(f"Last two issues : {issues[-2:]}")
print("-" * 40)

#sets

reporters = set()
tech_stack = set()

for issue in issues:
  reporters.add(issue["reporter"]) #adding the reporter to the set

for c in contributors:
  tech_stack.add(c["language"]) #adding the language to the set

#add tech manually
tech_stack.add("Python")
#discard
tech_stack.discard("Java")

union = reporters.union(tech_stack) #merge both sets
intersection = reporters.intersection(tech_stack) #common items
difference = reporters.difference(tech_stack)

#check criticals
priorities = set()
for issue in issues:
  priorities.add(issue["priority"]) #making a set of all priorities

if "Critical" in priorities:
  result = "YES - flag for immediate review"
else:
  result = "NO"

#print
print(f"reporters         : {reporters}")
print(f"tech_stack        : {tech_stack}")
print(f"union             : {union}")
print(f"intersection      : {intersection}   difference  :{difference}")
print(f"Critical present  : {result}")
print("-" * 40)


#priority count dict

priority_count = {}

for issue in issues:
  prio = issue["priority"]

  if prio in priority_count:
    priority_count[prio] += 1 #keep adding if theres more
  else:
    priority_count[prio] = 1 #add 1 if seen for the first time



#status_groups (grouping by status (open, resolved))
status_groups = {}

for issue in issues:
  stat = issue["status"]  #hold the status

  if stat not in status_groups:
    status_groups[stat] = [] #create new key-value pair in dict

  status_groups[stat].append(issue)  #append the issue as value of the status key



#use function

print(f"priority keys    : dict_keys({priority_count.keys()})")
print(f"priority values  : dict_values({priority_count.values()})")

result = [] #make a list to combine the status group

for k, v in status_groups.items():
    result.append(f"{k} : {v}")

print(f"Status Groups → {' | '.join(result)}")

#top reporter

report_count = {}  #create {"reporter name (r)" : report count (count)}

for issue in issues:
  r = issue["reporter"] #hold the reporter

  if r in report_count:
    report_count[r] += 1
  else:
    report_count[r] = 1 #when appear for the 1st time


top_r = None
max_count = 0

for r, count in report_count.items():
  if count > max_count:
    max_count = count
    top_r = r

print(f"Top Reporter    : {top_r} ({max_count} issues)")

# Use pop() to remove the 'type' key from the first issue
# Fix: Use pop() with a default value to avoid KeyError if the key is not present.
issues[0].pop('types')
print("After pop('type'):", issues[0])
print("-" * 40)

#SECTION 3 SAVE READ AND REPORT

import os

folder_name = project[0].lower().replace(" ","-")

if not os.path.exists(folder_name):
  os.mkdir(folder_name) #create folder
  print(f"Folder created  : {folder_name}")


#Use os.path.join() for all paths — print folder contents with os.listdir()
report_path = os.path.join(folder_name, "project_report.txt")
csv_path = os.path.join(folder_name, "contributors.csv")


#save files
try:
  with open(report_path, "w") as f:
    f.write("=" * 40 + "\n")
    f.write(f"{project[0]} — PROJECT REPORT\n")
    f.write("=" * 40 + "\n")
    f.write(f"Project Info : {project}\n")
    f.write(f"Contributors : {contributors}\n")
    f.write(f"Issues       : {issues}\n")
    f.write(f"Top Reporter : {top_r} ({max_count} issues)\n")



  with open(csv_path, "w") as f:
    f.write("id,title,priority,reporter,status\n")

    for issue in issues:
        f.write(f"{issue['id']},{issue['title']},{issue['priority']},{issue['reporter']},{issue['status']}\n")

  print(f"Files saved : {os.listdir(folder_name)}")

except IOError:
  print("Error writing files.")




#read files
try:
    # read()
    with open(report_path, "r") as f:
        print("\n--- read() ---")
        print(f.read())  #read the whole content

    # readline()
    with open(report_path, "r") as f:
        print("--- readline() ---")
        print(f"Line 1 : {f.readline()}")
        print(f"Line 2 : {f.readline()}")

    # readlines()
    with open(report_path, "r") as f:
        lines = f.readlines()

        print("--- readlines() ---")

        count = 0
        for line in lines:
            if "Critical" in line or "High" in line:
                count += 1
        print(f"Total lines : {len(lines)}   Critical/High lines : {count}")
        print("-" * 40)
except FileNotFoundError:
    print("File not found")

#final summary

print("=" * 40)
print(f"{project[0]} — FINAL SUMMARY")
print("=" * 40)

print(f"Project : {project[0]}     Version : {project[1]}     Lead : {project[4]}")
print(f"Contributors : {len(contributors)}   Names : {names}")
print(f"Tech Stack   : {tech_stack}")
print(f"Issues  : {len(issues)}   Open : {open_count}   Reporters : {len(reporters)}")
print(f"Top Reporter : {top_r} ({max_count} issues)")

# Priority breakdown
for k, v in priority_count.items():
    print(f"{k}:{v}", end="  ")

print(f"\nReport  : {report_path}")
print(f"CSV     : {csv_path}")
print("=" * 40)
print(f"{project[0]} complete. Thank you for contributing to open source!")


```
